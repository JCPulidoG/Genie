## Estimación de $\pi$ por medio de la Simulación Monte-Carlo

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Parámetros de la simulación
n = 2000000

# Generar puntos aleatorios en el cuadrado [-1, 1] x [-1, 1]
x = np.random.uniform(-1, 1, n)
y = np.random.uniform(-1, 1, n)

# Calcular el número de puntos dentro del círculo
inside_circle = (x**2 + y**2) <= 1
num_inside_circle = np.sum(inside_circle)
```

```python
# Estimar el valor de pi
pi_estimate = (num_inside_circle / n) * 4

print(f"Estimación de π usando {n} puntos: {pi_estimate}")
```

```python
# Visualizar los resultados
plt.figure(figsize = (6, 6))
plt.scatter(x[inside_circle], y[inside_circle], color = 'blue', s = 1)
plt.scatter(x[~inside_circle], y[~inside_circle], color = 'red', s = 1)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Estimación de π usando Monte Carlo')
plt.gca().set_aspect('equal', adjustable = 'box')
plt.show()
```
