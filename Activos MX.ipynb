```python
# Obtener todos los stocks disponibles de México
# Ejecutar UNA SOLA VEZ. TARDA 10 MINUTOS. 589 REGISTROS

import investpy as inv
import pandas as pd
from Functions import *

Country = "mexico"
Year_Start_Train, Year_End_Train, Year_Start_Test, Year_End_Test = 2010, 2019, 2020, 2020
Date_Start_Train, Date_End_Train, Date_Start_Test, Date_End_Test = fnGetDay(Year_Start_Train, "First"), fnGetDay(Year_End_Train, "Last"), fnGetDay(Year_Start_Test, "First"), fnGetDay(Year_End_Test, "Last")

# Obtener todos los stocks disponibles de México
stocks_mx = inv.stocks.get_stocks(country = Country)
stocks_mx = stocks_mx["symbol"] + ".MX"
stocks_mx = stocks_mx.drop_duplicates()
Columns = ["Symbol", "Price_Start", "Price_End"]
Prices = []
I = 1

# Obtener los Rendimientos de todos los Stocks de México
for Stock in stocks_mx:
    try:
        Array_Prices = []
        Price_Start = round(fnGetPrice(Stock, Date_Start_Train), 4)
        Price_End = round(fnGetPrice(Stock, Date_End_Train), 4)
        Array_Prices = [Stock, Price_Start, Price_End]
        Prices.append(Array_Prices)
        print(I)
        I += 1
    except Exception as e:
        print(f"No se encontró el stock {Stock} o ocurrió un error: {e}")
        print(I)
        I += 1
        continue

df_Stocks = pd.DataFrame(Prices, columns = Columns)
df_Stocks = df_Stocks.drop(df_Stocks[df_Stocks['Price_Start'] == 0].index)
df_Stocks = df_Stocks.drop(df_Stocks[df_Stocks['Price_End'] == 0].index)
df_Stocks = df_Stocks.reset_index(drop = True)
df_Stocks.to_csv("MexicanStocks10Years.csv")
```
